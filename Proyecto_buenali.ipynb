```python
def distancia_una(): 
    parejas = list(combinations(range(100),2))
    pareja = choice(parejas)
    seq1 = seqs[pareja[0]]
    seq2 = seqs[pareja[1]]
    conast1 = str(seq1.seq)
    conast2 = str(seq2.seq)
    if len(conast1) == len(conast2):
        return 1000*hamming(Sequence(conast1),Sequence(conast2))
```

```python
distancia_una()
```

```python
mil_veces = [distancia_una() for _ in range(1000)]
mil_veces2 = [x for x in mil_veces if x is not None]
```

```python
import numpy as np
import matplotlib.pyplot as plt

counts, bins = np.histogram(mil_veces2, bins=50)
plt.stairs(counts, bins)
plt.title("Histograma")
```
